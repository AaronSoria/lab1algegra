```python
import numpy as np
import cmath
import math


def mostrar_valores(nombre, det, magnitud, fase, traza):
    print(f"Matriz: {nombre}")
    print(f"Determinante: {det}")
    print(f"Módulo (cmath): {magnitud}")
    print(f"Fase: {fase} radianes")
    print(f"Forma polar: {magnitud} * e^(i*{fase})")
    print(f"Traza: {traza}")

A = np.array([
    [3 + 2j, 1 - 1j, 4 - 3j],
    [2 + 1j, -5 + 4j, 6 + 0j],
    [-3 + 2j, 7 - 1j, 8 + 3j]
],  dtype=complex)

B = np.array([
    [2+3j, 4+0j, 1+2j],
    [3-2j, 5+1j, 7-3j],
    [6+4j, -2+0j, 9+5j],
],  dtype=complex)

C = np.array([
    [2+3j, 4+0j, 1+2j, -5+0j],
    [-2+3j, 6+0j, 4+7j, 2+1j],
    [-1+2j, -6+5j, 0-8j, 7+4j],
    [-3-4j, 3-1j, -5-3j, 0+5j],
],  dtype=complex)

D = np.array([
    [0+1j, 2+0j],
    [2-3j, 4+5j]
],  dtype=complex)
```

```python
detA = np.linalg.det(A)
magnitudA = cmath.polar(detA)[0]
faseA = cmath.polar(detA)[1] #en rad
trazaA = np.trace(A)
mostrar_valores("A",detA,magnitudA,faseA, trazaA)

detB = np.linalg.det(B)
magnitudB = cmath.polar(detB)[0]
faseB = cmath.polar(detB)[1]
trazaB = np.trace(B)
mostrar_valores("B",detB,magnitudB,faseB, trazaB)

detC = np.linalg.det(C)
magnitudC = cmath.polar(detC)[0]
faseC = cmath.polar(detC)[1]
trazaC = np.trace(C)
mostrar_valores("C",detC,magnitudC,faseC, trazaC)

detD = np.linalg.det(D)
magnitudD = cmath.polar(detD)[0]
faseD = cmath.polar(detD)[1]
trazaD = np.trace(D)
mostrar_valores("D",detD,magnitudD,faseD, trazaD)


resultado = (0+2j)*(detD+trazaC)*(A @ B) - (detC*trazaD)*(A @ B)
print("resultado")
print(resultado)
```

```python
inv_A = np.linalg.inv(A)
inv_B = np.linalg.inv(B)
inv_C = np.linalg.inv(C)
inv_D = np.linalg.inv(D)

print("inversa de A")
print(inv_A)
print("inversa de B")
print(inv_B)
print("inversa de C")
print(inv_C)
print("inversa de D")
print(inv_D)
```

```python
eigenvaluesA, _ = np.linalg.eig(A)
eigenvaluesB, _ = np.linalg.eig(B)
eigenvaluesC, _ = np.linalg.eig(C)
eigenvaluesD, _ = np.linalg.eig(D)
```

```python
from qiskit import QuantumCircuit
from qiskit.primitives import StatevectorSampler
from qiskit.visualization import plot_histogram
 
qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0, 1)
qc.measure_all()
 
sampler = StatevectorSampler()
result = sampler.run([qc], shots=1024).result()
counts = result[0].data.meas.get_counts()
print(counts)
plot_histogram(counts) ## grafica un estado de bell 00 o 11
```
