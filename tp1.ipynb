```python
import numpy as np
import cmath
import math


def mostrar_valores(nombre, det, magnitud, fase, traza, magnitudTraza, faseTraza):
    print(f"Matriz: {nombre}")
    print(f"Determinante: {det}")
    print(f"Módulo (cmath): {magnitud}")
    print(f"Fase: {fase} radianes")
    print(f"Forma polar: {magnitud} * e^(i*{fase})")
    print(f"Traza: {traza}")
    print(f"Forma polar: {magnitudTraza} * e^(i*{faseTraza})")

def es_hermitiana(M):
    return np.allclose(M, M.conjugate().T)

def es_normal(M):
    return np.allclose(M @ M.conjugate().T, M.conjugate().T @ M)

def es_unitaria(M):
    I = np.eye(M.shape[0])
    return np.allclose(M @ M.conjugate().T, I)

A = np.array([
    [3 + 2j, 1 - 1j, 4 - 3j],
    [2 + 1j, -5 + 4j, 6 + 0j],
    [-3 + 2j, 7 - 1j, 8 + 3j]
],  dtype=complex)

B = np.array([
    [2+3j, 4+0j, 1+2j],
    [3-2j, 5+1j, 7-3j],
    [6+4j, -2+0j, 9+5j],
],  dtype=complex)

C = np.array([
    [2+3j, 4+0j, 1+2j, -5+0j],
    [-2+3j, 6+0j, 4+7j, 2+1j],
    [-1+2j, -6+5j, 0-8j, 7+4j],
    [-3-4j, 3-1j, -5-3j, 0+5j],
],  dtype=complex)

D = np.array([
    [0+1j, 2+0j],
    [2-3j, 4+5j]
],  dtype=complex)
```

```python
matrices = {"A": A, "B": B, "C": C, "D": D}

for nombre, M in matrices.items():
    print(f"\n---- Matriz {nombre} ----")
    print("Hermitiana:", es_hermitiana(M))
    print("Normal    :", es_normal(M))
    print("Unitaria  :", es_unitaria(M))
```

```python
for nombre, M in matrices.items():
    detM = np.linalg.det(M)
    magnitudM = cmath.polar(detM)[0]
    faseM = cmath.polar(detM)[1] #en rad
    trazaM = np.trace(M)
    magnitudMTraza = cmath.polar(trazaM)[0]
    faseMTraza = cmath.polar(trazaM)[1] #en rad
    mostrar_valores(nombre, detM,magnitudM,faseM, trazaM, magnitudMTraza, faseMTraza)
```

```python
detD = np.linalg.det(D)
trazaC = np.trace(C)
detC = np.linalg.det(C)
trazaD = np.trace(D)
resultado = (0+2j)*(detD+trazaC)*(A @ B) - (1+1j)*(detC*trazaD)*(A @ B)
print("2i(det(D)+tr(C))AB-(1+i)*det(C)tr(D)BA = ")
print(resultado)
```

```python
inv_A = np.linalg.inv(A)
inv_B = np.linalg.inv(B)
inv_C = np.linalg.inv(C)
inv_D = np.linalg.inv(D)

print("inversa de A")
print(inv_A)
print("inversa de B")
print(inv_B)
print("inversa de C")
print(inv_C)
print("inversa de D")
print(inv_D)
```

```python
for nombre, M in matrices.items():
    eigenvaluesM, _ = np.linalg.eig(M)
    print(f"valores propios de {nombre}")
    print(eigenvaluesM)
```

```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector

ket0 = np.array([1, 0])
ket1 = np.array([0, 1])


## encontrar ∣+⟩ 
qc = QuantumCircuit(1)
qc.initialize(ket0, 0)
qc.h(0)

out_vector = Statevector(qc)
print(out_vector)
 
# dibujar esfera de bloch
plot_bloch_multivector(out_vector)
display(out_vector.draw('latex'))
```

```python
## encontrar ∣-⟩ 
qc = QuantumCircuit(1)
qc.initialize(ket0, 0)
qc.x(0)
qc.h(0)

out_vector = Statevector(qc)
print(out_vector)
 
# dibujar esfera de bloch
plot_bloch_multivector(out_vector)
display(out_vector.draw('latex'))
```

```python
## encontrar ∣+i⟩
qc = QuantumCircuit(1)
qc.initialize(ket0, 0)
qc.h(0)
qc.s(0)

out_vector = Statevector(qc)
print(out_vector)
 
# dibujar esfera de bloch
plot_bloch_multivector(out_vector)
display(out_vector.draw('latex'))
```

```python
## encontrar ∣-i⟩
qc = QuantumCircuit(1)
qc.initialize(ket0, 0)
qc.h(0)
qc.s(0)
qc.x(0)


out_vector = Statevector(qc)
print(out_vector)
 
# dibujar esfera de bloch
plot_bloch_multivector(out_vector)
display(out_vector.draw('latex'))
```

```python
from qiskit import QuantumCircuit
 ## base C4
ket00 = np.kron(ket0,ket0)
ket01 = np.kron(ket0,ket1)
ket10 = np.kron(ket1,ket0)
ket11 = np.kron(ket1,ket1)

kets = {"00": ket00, "01": ket01, "10": ket10, "11": ket11}

for _, ket in kets.items():
    qc = QuantumCircuit(2)
    qc.initialize(ket, [0, 1]) 
    state = Statevector.from_instruction(qc)
    display(state.draw('latex'))
    print(state) 
```
